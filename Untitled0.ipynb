```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure column names are clean (strip any leading/trailing whitespace)
# This was already done in the previous cell, but good for robustness if this cell is run independently.
df.columns = df.columns.str.strip()

# Convert 'Temperatura Media (C)' to numeric, handling comma decimals
# This was already done in the previous cell, but good for robustness if this cell is run independently.
df['Temperatura Media (C)'] = df['Temperatura Media (C)'].astype(str).str.replace(',', '.', regex=False).astype(float)

# Define features (X) and target (y)
X = df[['Temperatura Media (C)']]
y = df['Consumo de cerveja (litros)'] # Using the corrected column name

# Combine X and y and drop rows with any NaN values to ensure a clean dataset for regression
combined_data = pd.concat([X, y], axis=1).dropna()
X_cleaned = combined_data[['Temperatura Media (C)']]
y_cleaned = combined_data['Consumo de cerveja (litros)']

if X_cleaned.empty or y_cleaned.empty:
    print("Error: No valid data remaining after cleaning for linear regression. Check your dataset and column names.")
else:
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X_cleaned, y_cleaned, test_size=0.2, random_state=42)

    # Initialize and train the linear regression model
    model = LinearRegression()
    model.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = model.predict(X_test)

    # Evaluate the model
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    # Print the results
    print(f"Intercept: {model.intercept_:.2f}")
    print(f"Coefficient (Temperatura Media): {model.coef_[0]:.2f}")
    print(f"Mean Squared Error: {mse:.2f}")
    print(f"R-squared: {r2:.2f}")

    # Visualize the regression line
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=X_cleaned['Temperatura Media (C)'], y=y_cleaned, alpha=0.6)
    # Sort X_test to ensure the line is drawn correctly
    # Create a DataFrame from X_test for easier sorting and plotting
    X_test_df = pd.DataFrame(X_test, columns=['Temperatura Media (C)'])
    # Add predictions to the DataFrame for sorting
    X_test_df['y_pred'] = y_pred
    X_test_df_sorted = X_test_df.sort_values(by='Temperatura Media (C)')

    plt.plot(X_test_df_sorted['Temperatura Media (C)'], X_test_df_sorted['y_pred'], color='red', linewidth=2, label='Regression Line')
    plt.title('Regressão Linear: Temperatura Média vs. Consumo de Cerveja')
    plt.xlabel('Temperatura Média (C)')
    plt.ylabel('Consumo de Cerveja (litros)')
    plt.legend()
    plt.grid(True)
    plt.show()
```
