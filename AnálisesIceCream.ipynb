```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("raphaelmanayon/temperature-and-ice-cream-sales")

print("Path to dataset files:", path)
```

Carregamento de Dados

```python
import pandas as pd
import os

# O caminho para o diretório dos arquivos do dataset foi obtido na célula anterior
# A variável `path` contém este caminho.

# Nome do arquivo CSV que esperamos encontrar, baseado nas informações do kernel.
file_name = 'Ice Cream Sales - Temperature Data.csv'
full_path = os.path.join(path, file_name)

# Verificar se o arquivo existe no caminho esperado
if not os.path.exists(full_path):
    print(f"Aviso: O arquivo '{file_name}' não foi encontrado diretamente em '{path}'.")
    print("Listando o conteúdo do diretório para encontrar o nome correto do arquivo:")
    files_in_dir = os.listdir(path)
    for item in files_in_dir:
        print(f"- {item}")

    # Tentar encontrar o primeiro arquivo CSV na lista
    csv_files = [f for f in files_in_dir if f.lower().endswith('.csv')]
    if csv_files:
        # Se um CSV for encontrado, use-o
        detected_file_name = csv_files[0]
        full_path = os.path.join(path, detected_file_name)
        print(f"Usando o arquivo CSV detectado: '{detected_file_name}'")
        print(f"Caminho completo: {full_path}")
    else:
        # Se nenhum CSV for encontrado, levantar um erro
        raise FileNotFoundError(f"Nenhum arquivo CSV encontrado em '{path}'. Por favor, verifique o conteúdo do dataset.")

# Carregar os dados no DataFrame do Pandas
df = pd.read_csv(full_path)

# Exibir as primeiras linhas do DataFrame
print(f"DataFrame criado com sucesso a partir de '{os.path.basename(full_path)}'. Aqui estão as primeiras linhas:")
display(df.head(10))
```
